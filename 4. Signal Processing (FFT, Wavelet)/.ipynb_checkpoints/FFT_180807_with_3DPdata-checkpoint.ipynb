# 실험데이터(3DP) 이용 FFT 수행

### 라이브러리 불러오기

```python
import pandas            as pd
import numpy             as np
import math              as mat
import matplotlib        as mpl
import matplotlib.pyplot as plt
```

### 데이터 불러오기 (pandas DataFrame)

```python
rawData = pd.read_csv('./Data/Example_data_3DP_Full.txt', sep='\t', names=['time', 'ACCx', 'ACCy', 'ACCz', 'AE'])

print(rawData.shape)
print(rawData.info())
rawData
```

### ACC x축 데이터만 추출

```python
acc_x_data = rawData.iloc[:,1:2]
acc_x_data
```

### ACC x축 데이터 그래프 확인

### FFT 수행 

```python
Fs   = 6400
NFFT = len(acc_x_data)
k    = np.arange(NFFT)
f0   = k * Fs / NFFT
f0   = f0[range(mat.trunc(NFFT/2))]

fftResult    = np.fft.fft(acc_x_data) / NFFT
fftResult    = fftResult[range(mat.trunc(NFFT/2))]
fftMagnitude = 2 * abs(fftResult)
fftPhageAng  = np.angle(fftResult) * 180/np.pi
```

### FFT 결과 그래프 확인

```python
mpl.rcParams.update({'font.size': 5, 'grid.linestyle':':', 'grid.linewidth':0.4})

ticksfontsize = 5
titlefont = {'size': 8}
labelfont = {'size': 5}

plt.figure(num=2, dpi=200)
plt.subplots_adjust(hspace = 0.5, wspace = 0.3)

plt.subplot(2,1,1)
plt.plot(rawData.time, acc_x_data, lw=0.3)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0), useMathText=True)
plt.xticks(size = ticksfontsize)
plt.yticks(size = ticksfontsize)
plt.title('Time Domain'     , fontdict=titlefont) # "fontsize=8"로 써도 무방
plt.xlabel('time($sec$)'    , fontdict=labelfont)     # "fontsize=5"로 써도 무방
plt.ylabel('acceration($g$)', fontdict=labelfont)
plt.grid()

plt.subplot(2,1,2)
plt.plot(f0, fftMagnitude, 'r', lw=0.3)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0), useMathText=True)
plt.xticks(size = ticksfontsize)
plt.yticks(size = ticksfontsize)
plt.xlim([0, 1000])
plt.title('Frequency Domain (FFT)' , fontdict=titlefont) # "fontsize=8"로 써도 무방
plt.xlabel('Frequency($Hz$)'       , fontdict=labelfont)     # "fontsize=5"로 써도 무방
plt.ylabel('Magnitude'             , fontdict=labelfont)
plt.grid()

'''
plt.subplot(3,1,3)
plt.plot(f0, fftPhageAng, lw=0.3)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0), useMathText=True)
plt.xticks(size = ticksfontsize)
plt.yticks(size = ticksfontsize)
plt.title('Phage Angle'         , fontdict=titlefont) # "fontsize=8"로 써도 무방
plt.xlabel('Frequency($Hz$)'    , fontdict=labelfont)     # "fontsize=5"로 써도 무방
plt.ylabel('angle($deg.$)'      , fontdict=labelfont)
plt.grid()
'''

plt.savefig("./FFTgraph(3dp)/Fig2_FFTresult.png", dpi=300)
plt.show()
```
